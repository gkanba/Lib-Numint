```python
%matplotlib ipympl
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (8, 8))
ax = fig.add_subplot(111, projection='3d')

ax.set_title("Lorenz 63", size = 20)
ax.set_xlabel("x", size = 14)
ax.set_ylabel("y", size = 14)
ax.set_zlabel("z", size = 14)

#data = np.loadtxt("test.tsd", delimiter=",")
data = np.loadtxt("l63.tsd", delimiter=",")
data1 = np.delete(data, 0, 1)
data2 = np.delete(data1, 0, 1)

ax.plot(np.transpose(data2)[0], np.transpose(data2)[1], np.transpose(data2)[2], color="blue")
```
